```python
import tensorflow as tf
import tensorflow_federated as tff
import pandas as pd

# Function to load and convert data into TFF format
def load_dataset(csv_path):
    df = pd.read_csv(csv_path)

    # Separate features and labels
    X = df.drop(columns=["Label"])
    y = df["Label"]

    # Convert labels to 0s and 1s (binary classification)
    y = y.apply(lambda label: 1 if label != "BENIGN" else 0)

    # Convert to TensorFlow Dataset
    dataset = tf.data.Dataset.from_tensor_slices((X.values, y.values))

    # Batch the dataset
    dataset = dataset.batch(32)
    return dataset

# Load data for each organization
org1_dataset = load_dataset("org1_data.csv")
org2_dataset = load_dataset("org2_data.csv")
org3_dataset = load_dataset("org3_data.csv")

# Store in a list for Federated Learning
federated_train_data = [org1_dataset, org2_dataset, org3_dataset]

print("✅ Data successfully loaded for Federated Learning!")
```

```python
import tensorflow as tf
import tensorflow_federated as tff

# Define a simple model for network anomaly detection
def create_keras_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(64, activation='relu', input_shape=(78,)),  # Adjust input shape based on features
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')  # Binary classification (Benign vs. Attack)
    ])
    return model

# Convert Keras model to TFF format
def model_fn():
    keras_model = create_keras_model()

    # Define loss and metrics
    loss_fn = tf.keras.losses.BinaryCrossentropy()
    metrics = [tf.keras.metrics.BinaryAccuracy()]

    # Convert the Keras model into a Federated Learning model
    return tff.learning.models.from_keras_model(
        keras_model,
        input_spec=federated_train_data[0].element_spec,  # Ensure this is defined before use
        loss=loss_fn,
        metrics=metrics
    )

print("✅ Model successfully defined!")



# Create Federated Learning process
iterative_process = tff.learning.algorithms.build_weighted_fed_avg(
    model_fn,
    client_optimizer_fn=lambda: tf.keras.optimizers.Adam(learning_rate=0.01)
)

# Initialize FL state
state = iterative_process.initialize()

print("✅ Federated Learning process initialized!")
```

```python
NUM_ROUNDS = 10  # Adjust the number of training rounds

for round_num in range(1, NUM_ROUNDS + 1):
    state, metrics = iterative_process.next(state, federated_train_data)
    print(f"Round {round_num}: {metrics}")

print("✅ Federated Learning training complete!")
```

```python
# Load test data
test_dataset = load_dataset("test_data.csv")

# Convert dataset to TFF format
evaluation_process = tff.learning.algorithms.build_fed_eval(model_fn)

# Initialize the evaluation process
evaluation_state = evaluation_process.initialize()

# Update the evaluation state with the model weights from the training state
# Assuming `state` contains the model weights in a compatible format
evaluation_state = evaluation_process.set_model_weights(evaluation_state, state.global_model_weights)

# Evaluate the model on the test dataset
# The `next` method returns a tuple: (updated_state, metrics)
updated_state, test_metrics = evaluation_process.next(evaluation_state, [test_dataset])

# Access the accuracy from the metrics
# Check the structure of test_metrics to find the correct key
print("Test Metrics Structure:", test_metrics)

# Access binary_accuracy from current_round_metrics or total_rounds_metrics
if 'client_work' in test_metrics and 'eval' in test_metrics['client_work']:
    eval_metrics = test_metrics['client_work']['eval']
    if 'current_round_metrics' in eval_metrics:
        current_round_metrics = eval_metrics['current_round_metrics']
        if 'binary_accuracy' in current_round_metrics:
            accuracy = current_round_metrics['binary_accuracy']
            print("Test Binary Accuracy (Current Round):", accuracy)
        else:
            print("Key 'binary_accuracy' not found in current_round_metrics.")
    elif 'total_rounds_metrics' in eval_metrics:
        total_rounds_metrics = eval_metrics['total_rounds_metrics']
        if 'binary_accuracy' in total_rounds_metrics:
            accuracy = total_rounds_metrics['binary_accuracy']
            print("Test Binary Accuracy (Total Rounds):", accuracy)
        else:
            print("Key 'binary_accuracy' not found in total_rounds_metrics.")
    else:
        print("Neither 'current_round_metrics' nor 'total_rounds_metrics' found in evaluation metrics.")
else:
    print("Evaluation metrics not found in the expected structure.")
```
